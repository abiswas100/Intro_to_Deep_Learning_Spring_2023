```python
%matplotlib inline
from __future__ import print_function
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
from tqdm import tqdm

(ds_train, ds_test), ds_info = tfds.load(
    'mnist',
    split=['train[:80%]', 'test[80%:]'],
    shuffle_files=True,
    as_supervised=True,
    with_info=True,
)

train_ds = ds_train.shuffle(1024).batch(32)
val_ds = ds_test.shuffle(1024).batch(32)

# Define the first model
model1 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28, 1)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

# Compile the first model
model1.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Add early stopping to the first model
early_stopping1 = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)


# Train the first model
history1 = model1.fit(train_ds, epochs=10, validation_data=val_ds)

# Define the second model
model2 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28, 1)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

# Compile the second model
model2.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Add early stopping to the first model
early_stopping2 = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)


# Train the second model
history2 = model2.fit(train_ds, epochs=10, validation_data=val_ds)

# Define the third model
model3 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28, 1)),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

# Compile the third model
model3.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Add early stopping to the first model
early_stopping3 = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)

# Train the third model
history3 = model3.fit(train_ds, epochs=10, validation_data=val_ds)


```
